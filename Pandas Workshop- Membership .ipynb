# Slicing using Operators

```python
heart['age']>60
```

```python
heart[heart['age']>60]
```

```python
heart[(heart['age']>60) & (heart['sex']==1) & (heart['cp']==1)]
```

```python
heart[(heart['age']>60) | (heart['sex']==1) & (heart['cp']==1)]
```

```python
heart[(heart['age']>60) & (heart['sex']==1) | (heart['cp']==1)]
```
